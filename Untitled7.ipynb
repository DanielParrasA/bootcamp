```python
import torch
print(f"CUDA available: {torch.cuda.is_available()}")
```

```python
import os

output_dir = "/content/generated_images"
os.makedirs(output_dir, exist_ok=True)

print(f"Directorio de salida creado en: {output_dir}")
```

```python
from diffusers import StableDiffusionPipeline, EulerAncestralDiscreteScheduler
from PIL import Image
```

```python
model_id = "stabilityai/stable-diffusion-2-1-base"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
```

```python
if torch.cuda.is_available():
    pipe.to("cuda")

prompt = "Un perro de raza bulldog negro con collar rosa"

image = pipe(
    prompt,
    guidance_scale=7.5,  # Controla la adherencia al prompt
    num_inference_steps=55  # Aumentar para mejor calidad
).images[0]

image_path = os.path.join(output_dir, "test2.png")
image.save(image_path)

print(f"Imagen guardada en: {image_path}")
```

```python
from diffusers import StableDiffusionImg2ImgPipeline
import requests
from io import BytesIO
```

```python
variations_dir = os.path.join(output_dir, "variations")
os.makedirs(variations_dir, exist_ok=True)
```

```python
image_url = "https://blog.dogfydiet.com/wp-content/uploads/2024/02/como-manejar-la-tendencia-del-Bulldog-Frances-Negro-a-roncar-.jpg"
response = requests.get(image_url)
image_base = Image.open(BytesIO(response.content)).convert("RGB")
image_base = image_base.resize((512, 512))
```

```python
base_image_path = os.path.join(output_dir, "base_image.png")
image_base.save(base_image_path)
print(f"✅ Imagen base guardada en: {base_image_path}")
```

```python
pipe_img2img = StableDiffusionImg2ImgPipeline.from_pretrained(
    model_id, torch_dtype=torch.float16
)
```

```python
pipe_img2img.to("cuda" if torch.cuda.is_available() else "cpu")

prompt_variation = "esta imagen de un perro con un fondo de navidad"
```

```python
image_variation = pipe_img2img(
    prompt=prompt_variation,
    image=image_base,
    strength=0.5,
    guidance_scale=7.5,
    num_inference_steps=50
).images[0]

variation_image_path = os.path.join(variations_dir, "variation_1.png")
image_variation.save(variation_image_path)
print(f"✅ Variación guardada en: {variation_image_path}")
```
